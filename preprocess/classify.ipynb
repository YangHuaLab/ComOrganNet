```python
import os
import shutil
import cv2
import numpy as np
```

```python
def mkdir(path):
    folder = os.path.exists(path)
    if not folder:
        os.makedirs(path)
```

```python
# Original filename format example: r02c04f03p01-ch1sk1fk1fl1.tiff, where ch1 indicates the image channel.
# Use this code to split the mixed image folder containing ch1-ch4 channels into 4 separate folders(ch1, ch2, ch3, ch4) under the whole_images directory.
```

```python
mkdir('./angii_drug_image/whole_images')
for ch in range(1, 5): 
    mkdir(f'./angii_drug_image/whole_images/ch{ch}')
```

```python
mkdir('./iso_drug_image/whole_images')
for ch in range(1, 5): 
    mkdir(f'./iso_drug_image/whole_images/ch{ch}')
```

```python
# 6.19-1 Example of data from different dates
prefix = "061901_"
src = 'E:/AXC-2022-6.19-1__2022-06-19T17_34_31-Measurement 1/Images/'
dsc = './angii_drug_image/whole_images/'

# Dosing images: columns 4-11 (angii data in columns 4-11 of the board)
for r in range(2, 8):
    for c in range(4, 12):
        for f in range(1, 12):
            for ch in range(1, 5):
                filename = f"r{'%02d'%r}c{'%02d'%c}f{'%02d'%f}p17-ch{ch}sk1fk1fl1.tiff"
                filename2 = prefix + filename
                if os.path.exists(os.path.join(src, filename)):
                    shutil.copyfile(os.path.join(src, filename), os.path.join(dsc, f"ch{ch}", filename2))
                else:
                    print(f"{filename2} doesn't exist")
                    missing_images.append(filename2)
```
