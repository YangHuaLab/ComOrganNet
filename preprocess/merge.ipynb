```python
# images of 4 channels(ch1,ch2,ch3,ch4) are merged to obtain the merge image
mkdir('./angii_drug_image/crop_images/crop_merge_ch1ch2ch3ch4')

fns1 = os.listdir("./angii_drug_image/crop_images/ch1_crop/")
fns2 = os.listdir("./angii_drug_image/crop_images/ch2_crop/")
fns3 = os.listdir("./angii_drug_image/crop_images/ch3_crop/")
fns4 = os.listdir("./angii_drug_image/crop_images/ch4_crop/")

fns1.sort()
fns2.sort()
fns3.sort()
fns4.sort()

for i in range(len(fns1)):
    ch1 = cv2.imread(f'./angii_drug_image/crop_images/ch1_crop/{fns1[i]}', cv2.IMREAD_UNCHANGED)
    ch2 = cv2.imread(f'./angii_drug_image/crop_images/ch2_crop/{fns2[i]}', cv2.IMREAD_UNCHANGED)
    ch3 = cv2.imread(f'./angii_drug_image/crop_images/ch3_crop/{fns3[i]}', cv2.IMREAD_UNCHANGED)
    ch4 = cv2.imread(f'./angii_drug_image/crop_images/ch4_crop/{fns4[i]}', cv2.IMREAD_UNCHANGED)

    ch_merge = cv2.merge([ch1, ch2, ch3, ch4])
    cv2.imwrite(f'./angii_drug_image/crop_images/crop_merge_ch1ch2ch3ch4/{fns1[i][:20]}{fns1[i][-11:]}', ch_merge)
```
